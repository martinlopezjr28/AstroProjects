```python
mdot_list = OS.read_hdf5(mdot_file)
```

#### define mdot attributes

```python
dmdt_BH1_array = mdot_list[6]
dmdt_BH2_array = mdot_list[7]
```

#### smooth mdot curves

```python
nbin = 75

table_smooth_BH1 = OS.bin_dmdt(dmdt_BH1_array,nbin)
log_numsnap_smooth_BH1 = table_smooth_BH1['numsnap_smooth']
dmdt_smooth_BH1 = table_smooth_BH1['dmdt_smooth']
M_integ_BH1 = table_smooth_BH1['M_integ']

table_smooth_BH2 = OS.bin_dmdt(dmdt_BH2_array,nbin)
log_numsnap_smooth_BH2 = table_smooth_BH2['numsnap_smooth']
dmdt_smooth_BH2 = table_smooth_BH2['dmdt_smooth']
M_integ_BH2 = table_smooth_BH2['M_integ']

numsnap_smooth_BH1 = 10**(log_numsnap_smooth_BH1)
numsnap_smooth_BH2 = 10**(log_numsnap_smooth_BH2)
```

#### integrated accreted mass

```python
M_accr1 = M_integ_BH1[0] * Mo
M_accr2 = M_integ_BH2[0] * Mo

print 'Integrated accreted mass for BH1:', round(M_accr1/Mo,2), 'Mo'
print 'Integrated accreted mass for BH2:', round(M_accr2/Mo,2), 'Mo\n'
```

#### disruption characteristics

```python
# TDE snapshots
TDE1_true = days_per_tdyn * 48
TDE2_true = days_per_tdyn * 278
```

#### plot mdot [Figure 4a]

```python
# mdot figure
mdot_fig = plt.figure(figsize=(40,22))

# setting axis
mdot_ax1 = mdot_fig.add_subplot(1,1,1)

# figure parameters
lw_size = 8
alpha_size = 0.5
label_size = 100
text_size = 75
tick_label_size = 60
major_tick_size = 24
minor_tick_size = major_tick_size/2
tick_width = 2

# conversions
true_t_days_BH1 = numsnap_smooth_BH1 * days_per_tdyn
true_t_days_BH2 = numsnap_smooth_BH2 * days_per_tdyn

# BH1
mdot_ax1.plot(true_t_days_BH1, dmdt_smooth_BH1,
              lw=lw_size, linestyle='--',
              label=r'$\rm{BH}_1$, $\rm{R}_{\rm{f}} = $' + str(round(mdot_radius,1)) + r'$\rm{R}_{\odot}$')
# BH2
mdot_ax1.plot(true_t_days_BH2, dmdt_smooth_BH2,
              lw=lw_size, label=r'$\rm{BH}_2$, $\rm{R}_{\rm{f}} = $' + str(round(mdot_radius,1)) + r'$\rm{R}_{\odot}$')

# TDE1
mdot_ax1.vlines(TDE1_true, 1e-10, 1e-5, lw= lw_size, alpha=alpha_size)
mdot_ax1.text(TDE1_true + 0.02, 5e-8,'t = ' + str(round(TDE1_true,2)) + ' days',
              size=text_size)
# TDE2
mdot_ax1.vlines(TDE2_true, 1e-10, 1.5e-5, lw=lw_size, alpha=alpha_size)
mdot_ax1.text(TDE2_true - 10.75, 5e-10,'t = ' + str(round(TDE2_true,2)) + ' days',
              size=text_size)

# x-axis
mdot_ax1.set_xlabel('time [days]', size=label_size)
mdot_ax1.set_xlim(np.min(true_t_days_BH2 - 1e-3),
                  np.max(true_t_days_BH2))
mdot_ax1.xaxis.set_ticks([1e0, 1e1])
mdot_ax1.loglog()
mdot_ax1.xaxis.set_tick_params(which='major', direction = 'out',
                               size=major_tick_size, width=tick_width,
                               labelsize=tick_label_size)
mdot_ax1.xaxis.set_tick_params(which='minor', direction='out',
                               size=minor_tick_size, width=tick_width)

# y-axis
mdot_ax1.set_ylabel(r'$\dot{\rm{M}}$ [$\rm{M}_{\odot}/\rm{s}$]', size=label_size)
mdot_ax1.set_ylim(np.min(dmdt_smooth_BH2),5e-6)
mdot_ax1.yaxis.set_ticks([1e-10, 1e-9, 1e-8, 1e-7, 1e-6, 1e-5])
mdot_ax1.yaxis.set_tick_params(which='major', direction = 'out',
                               size=major_tick_size, width=tick_width,
                               labelsize=tick_label_size)
mdot_ax1.yaxis.set_tick_params(which='minor', direction='out',
                               size=minor_tick_size, width=tick_width)

# legend
mdot_ax1.text(35, 5.0e-6,'OS', size=text_size)
mdot_ax1.legend(bbox_to_anchor=(0.35, 0.775), ncol=1, frameon=False, fontsize=text_size)

# figure properties
mdot_fig.tight_layout(pad=1.5, w_pad=1.6, h_pad=1.3)

# save figure
# mdot_fig.savefig('Figures/OS/Figure4a.png')
# mdot_fig.savefig('Figures/OS/Figure4a.eps')
# mdot_fig.savefig('Figures/Publication/Figure4/Figure4a.eps')
```
